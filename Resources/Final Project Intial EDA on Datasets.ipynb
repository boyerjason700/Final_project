```python

# Initial imports
import pandas as pd
from path import Path
```

```python
# Load the stats.csv dataset.
csv_path = Path("stats.csv")
stats_df = pd.read_csv(csv_path)
stats_df.head()
```

```python
#Process stats df 
stats_df['abbr'] = stats_df['Tm'].str.slice(stop=3)
stats_df['Year'] = stats_df['Season'].str.slice(start=5)
stats_df.head()
```

```python
# remove usless columns
stats_df = stats_df.drop(['Rk', 'Season', 'Tm', 'Lg'], axis=1)
stats_df.head()
```

```python

# Load the champions.csv dataset.
csv_path = Path("champions.csv")
champions_df = pd.read_csv(csv_path)
champions_df
```

```python
#Sort Champions df to only show Champions after 1990
champions_df = champions_df[(champions_df['Year'] > 1990)]
champions_df
```

```python

# Process champions df
champions_df['teamname'] = champions_df['Champion']
champions_df = champions_df.drop(['Champion','Lg', 'Runner-Up', 'Finals MVP', 'Unnamed: 5', 'Points', 'Rebounds', 'Assists', 'Win Shares'], axis=1)
champions_df
```

```python
# Load the team_name.csv dataset.
csv_path = Path("team_name.csv")
team_name_df = pd.read_csv(csv_path)
team_name_df
```

```python
champions_df['Year'] = champions_df['Year'].astype(str).str.slice(start=2)
#champions_df = champions_df.drop(['teamname'], axis = 1)
champions_df
```

```python
# merge Champions with thier stats from championship year- (PERFROM ML ON THIS DATA SET)
champ_stats_df = pd.merge(champions_df, stats_df, on=['Year', 'abbr'])
champ_stats_df
```
